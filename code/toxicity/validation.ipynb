```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
import seaborn as sns
import random
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
```

```python
df = pd.read_csv("/kaggle/input/reddit-perspective/comments_scores.csv")
```

```python
df
```

```python
df.iloc[1]["body"]
```

```python
threshold = 0.7
df['is_toxic'] = df['toxicity'] > threshold
df[df['toxicity'] > threshold]
```

```python
# Sample 25 true and 25 false rows
sample_true = df[df['is_toxic'] == True].sample(25, random_state=1)
sample_false = df[df['is_toxic'] == False].sample(25, random_state=1)

# Combine the samples
sample = pd.concat([sample_true, sample_false]).sample(frac=1, random_state=1).reset_index(drop=True)
sample['human_label'] = None
sample
```

```python
# Interactive labeling loop
for index, row in sample.iterrows():
    print(f"Comment {index+1}: {row['body']}")
    label = input("Is this toxic? (yes/no): ").strip().lower()
    sample.at[index, 'human_label'] = True if label == 'yes' else False
    
    
sample.to_csv("validation_sample.csv", index = False)
```

```python
sample['human_label'] = sample['human_label'].astype(bool)
sample['is_toxic'] = sample['is_toxic'].astype(bool)
sample
```

```python
# Calculate the metrics
accuracy = accuracy_score(sample['is_toxic'], sample['human_label'])
f1 = f1_score(sample['is_toxic'], sample['human_label'])
precision = precision_score(sample['is_toxic'], sample['human_label'])
recall = recall_score(sample['is_toxic'], sample['human_label'])
```

```python
# Code to print the performance metrics with text
print(f"Performance Metrics based on the provided labels:")
print(f"Accuracy: {accuracy:.2f}")
print(f"F1 Score: {f1:.3f}")
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
```
